# Cleaning RansomHub Data

```python
df = pd.read_csv('CSV_files/RansomHub.csv')
df.loc[-1] = ['2024-09-29', 'markdom.com', 'RansomHub']
df.index = df.index + 1
df.sort_index(inplace=True)
df.columns = ['Date', 'Victim', 'Ransomware Group']
df.insert(2, "Country", "")
```

```python
import ollama
index = 0
for victim in df['Victim']:
    response = ollama.chat(model='llama3', messages=[
        {
            'role': 'user',
            'content': f'I am going to give you the website of a victim of ransomware. I need you to provide me with the country the company is based in. Nothing else. Just the country. No other information. Give only the sountry as your response. The victim site is {victim}',

        },
    ])

    country = (response['message']['content'])
    print(country)

    rowIndex = df.index[index]
    df.loc[rowIndex, 'Country'] = country
    
    index += 1
```

```python
df['Country'].replace('USA', 'United States', inplace=True)
df['Country'].replace('UK', 'United Kingdom', inplace=True)
df['Date'] = pd.to_datetime(df['Date'], format="%Y %m %d")
```

```python
df.to_csv('RansomHub_Cleaned.csv')
```

# Distribution of Countries Targetted by Group

## Countries Targetted by RansomHub

```python
rh = pd.read_csv('CSV_files/Final_Distribution_Files/RansomHub_Cleaned.csv')
percentages = rh['Country'].value_counts(normalize=True).mul(100).round(1).astype(str) + '%'
countries = pd.DataFrame(rh['Country'])
countries = countries.drop_duplicates()
countries = countries.reset_index(drop=True)
countries.insert(1, column='Percent', value=0)

for i in range(len(countries)):
    country = countries['Country'][i]
    percent = percentages[country]
    countries['Percent'].iloc[i] = f'{percent}'

countries.to_csv("RansomHub_Countries.csv")
```
